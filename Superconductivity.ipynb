```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('train.csv')
```

```python
df.head()
```

```python
df_x=df.iloc[:,0:80]
df_y=df.iloc[:,81]
```

```python
df_x.head()
```

```python
df_y.head()
```

```python
df.describe().T
```

```python
from sklearn.preprocessing import StandardScaler
x = StandardScaler().fit_transform(df_x)
x
```

```python
from sklearn.decomposition import PCA
pca=PCA(n_components=9)
principalcondn=pca.fit_transform(x)
```

```python
pca.explained_variance_ratio_.cumsum()
```

```python
from sklearn.tree import DecisionTreeRegressor

DT = DecisionTreeRegressor(max_depth=17)
DT.fit (df_x,df_y)
```

```python
predicted_labels= DT.predict(df_x)
```

```python
from sklearn.metrics import mean_squared_error
from math import sqrt

rms = sqrt(mean_squared_error(df_y, predicted_labels))
```

```python
print(rms)
```

```python
X=principalcondn
DT = DecisionTreeRegressor(max_depth=17)
DT.fit (X,df_y)
```

```python
X.shape,df_y.shape
```

```python
predicted_labels= DT.predict(X)
```

```python
rms = sqrt(mean_squared_error(df_y, predicted_labels))
```

```python
print(rms)
```
